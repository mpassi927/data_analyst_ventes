# Découverte des données

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#création d'une liste des noms des fichiers
files=[file for file in os.listdir(r'C:/Users/FRANCHOMME LAETITIA/3D Objects/mes_fichiers_de_base')]
for file in files:
    print(file)
```

```python
path=r'C:/Users/FRANCHOMME LAETITIA/3D Objects/mes fichiers de base'
#création de la base de donnée vide
all_data=pd.DataFrame()
#base de donnée intermédiaire
for file in files:
    current_data=pd.read_csv(path+'/'+file)
    all_data=pd.concat([all_data,current_data])
    print(all_data)
```

```python
all_data.to_csv(path+'/all_data', index=False)
```

```python
#Voir les valeurs manquqntes
all_data.isnull().sum()
```

```python
#Supprimer les valeurs manquqntes
all_data=all_data.dropna(how="all")
all_data.shape
```

```python
all_data.head()
```

# Quel est le mois durant lequel nous avons réalisé le meilleur chiffre d'affaire?

```python
#extraire les mois
def month(x):
    return x.split("/")[0]

all_data["month"]=all_data["Order Date"].apply(month)
all_data
```

```python
all_data.dtypes
```

```python
#Supprimer les order date dans la colonne month 
all_data=all_data[all_data["month"]!='Order Date']
all_data["month"].unique()
```

```python
# Transformation des types de données
all_data["Quantity Ordered"]=all_data["Quantity Ordered"].astype(int)
all_data["Price Each"]=all_data["Price Each"].astype(float)
all_data["month"]=all_data["month"].astype(int)
```

```python
all_data["chiffre d'affaire"]=all_data["Quantity Ordered"]*all_data["Price Each"]
all_data
```

```python
#regroupons par rapport aux mois en sommant le chiffre d'affaire de chaque mois
cf=all_data.groupby("month")["chiffre d'affaire"].sum()
cf
```

```python
#visualisons ces données
months=range(1,13) #création de l'axe vertical (les mois)
plt.bar(months,cf)
plt.xticks(months)
plt.xlabel("les mois en chiffre")
plt.ylabel("chiffre d'affaire")
plt.title("Répartition du chiffre d'affaire suivant les mois")
plt.show()
```

```python
#REPONSE: C'est le mois de Decembre oü il y'a eu plus de réalisation du chiffre d'affaire
```

# Dans quelle ville nous avons réalisé un maximum de commandes?

```python
#Extraire toujours
def ville(a):
    return a.split(",")[1]

all_data["Ville"]=all_data["Purchase Address"].apply(ville)
all_data
```

```python
 all_data.groupby("Ville")["Ville"].count()
```

```python
#visualisons
plt.bar(all_data.groupby("Ville")["Ville"].count().index,all_data.groupby("Ville")["Ville"].count().values)
plt.xticks(rotation="vertical")
plt.xlabel("les villes")
plt.ylabel("montant des commandes")
plt.title("Répartition des motants réalisés selon les villes")
plt.show()
```

```python
#REPONSE: C'est à San Francisco qu'il y'a eu plus de vente
```

# En quel moment doit on faire une compagne publicitaire pour avoir plus de ventes?

```python
all_data["heure"]=pd.to_datetime(all_data["Order Date"]).dt.hour
all_data
```

```python
keys=[]
hours=[]
for key,hour in all_data.groupby("heure"):
    keys.append(key)
    hours.append(len(hour))
hours
```

```python
plt.grid()
plt.plot(keys,hours)
plt.xlabel("les heures")
plt.ylabel("montants")
plt.title("Répartition des montants enregistrés par heure")
```

```python
#REPONSE: les meilleurs moments pour faire la compagne, c'est entre 12h et 19h
```
